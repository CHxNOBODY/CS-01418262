```python
data = data.rename(columns={"Area_in_Marla" : "area", "property_type" : "property"})
# .rename use to rename columns
data
```

```python
data["price"] = pd.to_numeric(data["price"], errors="coerce")
data["baths"] = pd.to_numeric(data["baths"], errors="coerce")
data["bedrooms"] = pd.to_numeric(data["bedrooms"], errors="coerce")
data["area"] = pd.to_numeric(data["area"], errors="coerce")

data.dtypes
```

```python
data
```

```python
data = data.dropna() # nan from to_numeric
data["baths"] = data["baths"].astype(int)
data["bedrooms"] = data["bedrooms"].astype(int)
data.dtypes
```

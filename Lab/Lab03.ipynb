```python
expensive_price = df_house["price"].quantile(0.4)
expensive_price
```

```python
def label_expensive(row):
    if row["price"] > expensive_price:
        return 1
    else:
        return 0
    
    # also we can write in this way
    # return 1 if row["price"] > expensive_price else 0
    
df_house["is_expensive"] = df_house.apply(label_expensive, axis=1)  # axis use to defind what do we want pandas to do with column
#  df_house["is_expensive"] = df_house.apply(lambda row: 1 if row["price"] > expensive_price else 0, axis=1)
df_house
```

```python
from pathlib import Path

df_images = {"image_path": [], "label": []}

for dir in Path("./cats_vs_dogs").iterdir():
    for path in dir.glob("*"):
        df_images["image_path"].append(path)
        df_images["label"].append(dir.name)
        
df_images = pd.DataFrame.from_dict(df_images)
df_images
```

```python
from PIL import Image

image_random = df_images.sample(n=1, random_state=123)["image_path"].iloc[0]
image_random = Image.open(image_random).convert("RGB")
image_random
```

```python
image_random.transpose(Image.FLIP_LEFT_RIGHT)
```

```python
image_random.transpose(Image.FLIP_TOP_BOTTOM)
```

```python
image_random.rotate(20)
```

```python
import numpy as np

img_array = np.array(image_random)
noise = np.random.uniform(0, 50, img_array.shape).astype("uint8")
image_noise = np.clip(img_array + noise, 0, 255)
image_noise = Image.fromarray(image_noise.astype(np.uint8))
image_noise
```
